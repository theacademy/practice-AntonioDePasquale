```python
import random
from enum import IntEnum

class Action(IntEnum):
    Rock = 1
    Paper = 2
    Scissors = 3

class RockPaperScissors:

    """Program to simulate playing the game of rock paper scissors vs a computer oponent"""

    def __init__(self):

        self.chosenHand = None
        self.computersHand = None

        self.tieCount = 0
        self.winCount = 0
        self.loseCount = 0

    """get the user's choice and return it as an Action enum value."""
    @staticmethod
    def usersChoice():

        while True:
            userInput = input("Choose your hand! (1 for rock, 2 for paper, 3 for scissors): ")

            if userInput in ['1', '2', '3']:
                return Action(int(userInput))
            
            else:
                print("please enter a valid choice (1 for rock, 2 for paper, or 3 for scissors)")

    """determine the winner of the round and update the counts."""
    def playSingleRound(self):
        self.chosenHand = self.usersChoice()
        self.computersHand = random.choice(list(Action))

        print(f"You selected: {self.chosenHand.name}")
        print(f"Your oponent selected: {self.computersHand.name}")

        result = self.determineWinner()
        print(result)

    """determine the winner of the game and print the results"""
    def determineWinner(self):

        if self.chosenHand == self.computersHand:
            self.tieCount += 1
            return "It's a tie!\n"
        
        elif (self.chosenHand == Action.Rock and self.computersHand == Action.Scissors) or \
             (self.chosenHand == Action.Scissors and self.computersHand == Action.Paper) or \
             (self.chosenHand == Action.Paper and self.computersHand == Action.Rock):
            self.winCount += 1
            return "You win!\n"
        
        else:
            self.loseCount += 1
            return "Computer wins!\n"

    """Play the game of rock paper scissors until the user decides to stop"""   
    def playRockPaperScissorsGame(self):
        while True:
            self.playSingleRound()
            playAgain = input("Do you want to play again? (y/n): ").lower()
            if playAgain!= "y":

                print("Thanks for playing, take a look at your scores!")
                print(f"Ties: {self.tieCount}, Wins: {self.winCount}, Losses: {self.loseCount}\n")

                self.tieCount = 0
                self.winCount = 0
                self.loseCount = 0

                break
        
    
"""Start the game"""
game = RockPaperScissors()
game.playRockPaperScissorsGame()
```
